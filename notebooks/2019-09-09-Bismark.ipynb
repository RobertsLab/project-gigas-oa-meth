```python
%%bash

find /Users/yaamini/Documents/project-gigas-oa-meth/analyses/2019-08-30-Bismark-Parameter-Testing/YRV*_R1_001.fastq.gz \
| xargs basename -s _R1_001.fastq.gz | xargs -I{} /Users/Shared/Apps/Bismark_v0.19.0/bismark \
--path_to_bowtie /Users/Shared/bioinformatics/bowtie2-2.2.9/ \
--samtools_path /Users/Shared/Apps/samtools-1.8/ \
--non_directional \
-p 4 \
-score_min L,0,-0.9 \
--genome /Users/yaamini/Documents/project-gigas-oa-meth/analyses/2019-08-30-Bismark-Parameter-Testing/Crassostrea_gigas.oyster_v9.dna_sm.toplevel/ \
-1 /Users/yaamini/Documents/project-gigas-oa-meth/analyses/2019-08-30-Bismark-Parameter-Testing/{}_R1_001.fastq.gz \
-2 /Users/yaamini/Documents/project-gigas-oa-meth/analyses/2019-08-30-Bismark-Parameter-Testing/{}_R2_001.fastq.gz
```

## 3. Deduplication

```python
%%bash

/Users/Shared/Apps/Bismark_v0.19.0/deduplicate_bismark \
--samtools_path /Users/Shared/Apps/samtools-1.8/ \
-p \
--bam \
YRV*_R1_001_bismark_bt2_pe.bam
```

## 4. Sorting for Downstream Applications

```python
%%bash

find *deduplicated.bam \
| xargs basename -s _R1_001_bismark_bt2_pe.deduplicated.bam | xargs -I{} /Users/Shared/Apps/samtools-1.8/samtools \
sort {}_R1_001_bismark_bt2_pe.deduplicated.bam \
-o {}_dedup.sorted.bam
```

## 5. Indexing for Downstream Applications

```python
%%bash

find *dedup.sorted.bam \
| xargs basename -s _dedup.sorted.bam | xargs -I{} /Users/Shared/Apps/samtools-1.8/samtools \
index {}_dedup.sorted.bam
```
